```python
import numpy as np
from matplotlib import pyplot
import time, sys

%matplotlib inline
```

```python
domain = 2
nx = 41
dx = domain / (nx - 1)
nt = 20
nu = 0.3
sigma = 0.2
dt = sigma * dx**2 / nu

u = np.ones(nx)
u[int(0.5 / dx):int((1 / dx) + 1)] = 2
print(dt)
```

```python
pyplot.plot(np.linspace(0, domain, nx), u)
pyplot.show()
```

```python
def centraldifference(nx):
    for i in range(nt):
        un = u.copy()
        u[1:-1] = un[1:-1] + (nu * dt / (dx)**2) * (un[2:] - 2*un[1:-1] + un[:-2])
    return u
```

```python
centraldifference(nx)
pyplot.plot(np.linspace(0, domain, nx), u)
pyplot.show()
```
