```python
import numpy as np
from matplotlib import pyplot
import time, sys

%matplotlib inline
```

```python
domain = 2
nx = 61         # number of spatial steps
nt = 20         # number of timesteps
c = 1           # wave propagation speed
CFL = 0.9       # Courant–Friedrichs–Lewy condition must be <= 1

dx = domain / (nx - 1)
dt = (dx * CFL) / c

u = np.ones(nx)
u[int(0.5 / dx):int((1 / dx) + 1)] = 2
print(dx)
print(dt)
```

```python
pyplot.plot(np.linspace(0, domain, nx), u);
```

```python
u_old = np.ones(nx) # temporary array to hold values at the previous timestep aka (at u_(n-1))

for n in range(nt):
    u_old = u.copy()
    u[1:] = u_old[1:] - (c * dt/dx) * (u_old[1:] - u_old[:-1])
```

```python
pyplot.plot(np.linspace(0, domain, nx), u);
```
